```python
import pandas as pd

from env import TSP
from agent import DQAgent
from runner import Runner
from net import FC
```

```python
df = pd.read_csv('trimmed_data.csv')
# df.drop_duplicates(subset = df.columns[-2:], keep = 'first', inplace = True)
data = df[['OUTLET_LATITUDE', 'OUTLET_LONGITUDE']].head(100).to_numpy()
print(df.info())
df.head()
```

```python
env = TSP(
    data = data, 
    # distance_method = 'euclidean_distance',
    fixed_start = True, 
    cum_reward = False, 
    render_mode = None
)
policy_net = FC(
    num_stops = data.shape[0], 
    hidden_dim = 128,
    output_dim = data.shape[0],
    mode = 'normal'
)
target_net = policy_net
agent = DQAgent(
    env = env, net = [policy_net, target_net],
    revert_discount = True,
    sample_type = 'Uniform',
    double = True, 
    device = 'cuda'
)
runner = Runner(env = env, agent = agent)
```

```python
num_episodes = 10_000

load = False
if load:
    runner.load_model(best = False)
    current_ep = len(runner.R_memory)
    if current_ep < num_episodes:
        num_episodes -= current_ep
        # num_episodes = num_episodes + 1

env, agent, best_R, best_route, training_chart, route_chart = runner.run(num_episodes = num_episodes, render_charts = [True]*2)
training_chart.show()
route_chart.show()
```
